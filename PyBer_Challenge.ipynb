### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

### total riders by city type

```python
#  1. Get the total rides for each city type

total_riders_by_city_type = pyber_data_df.groupby(['type']).count()['ride_id']
print(total_riders_by_city_type)
```

### total drivers by city type

```python
# 2. Get the total drivers for each city type (run not on merged pyber_df but city_data_df)

total_drivers_by_city_type = city_data_df.groupby(['type']).sum()['driver_count']
print(total_drivers_by_city_type)
```

### avg fare per rider = total fares/ total riders 

```python
#  4. In Step 4, calculate the average fare per ride by city type by dividing the sum of all the fares by the total rides.

# total fares 
total_fares_per_city_type = pyber_data_df.groupby(['type']).sum()['fare']
print(total_fares_per_city_type)

# avg fare = total fares/ total rides
average_riderfare_by_city_type = total_fares_per_city_type/total_riders_by_city_type
print(average_riderfare_by_city_type)
```

### avg fare per driver = total fares / total drivers

```python
# 5.In Step 5, calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers.
average_driverfare_by_city_type = total_fares_per_city_type/total_drivers_by_city_type
print(average_driverfare_by_city_type)
```

### Pyber summary DataFrame

```python
#  6. Create a PyBer summary DataFrame. 
# dictionary
pyber_summary = {
    'total rides' : total_riders_by_city_type,
    'total drivers' : total_drivers_by_city_type,
    'total fares' : total_fares_per_city_type,
    'average fare per ride' : average_riderfare_by_city_type,
    'average fare per driver' : average_driverfare_by_city_type}
# convert to dataframe
pyber_summary_df = pd.DataFrame(pyber_summary)
# show dataframe
pyber_summary_df
```

### remove index

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

### format pyber summary columns

```python
#  8. Format the columns.
pyber_summary_df['total fares'] = pyber_summary_df['total fares'].map('${:,.2f}'.format)
pyber_summary_df['average fare per ride'] = pyber_summary_df['average fare per ride'].map('${:,.2f}'.format)
pyber_summary_df['average fare per driver'] = pyber_summary_df['average fare per driver'].map('${:,.2f}'.format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

### DataFrame Created using groupby()

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_fare_by_DateAndCityType_df = pyber_data_df.groupby(['type','date']).sum()['fare']
sum_fare_by_DateAndCityType_df
```

### DataFrame Created using Pivot()

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
sum_fare_by_DateAndCityType_df = sum_fare_by_DateAndCityType_df.reset_index()
sum_fare_by_DateAndCityType_df
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_fares_pivot = sum_fare_by_DateAndCityType_df.pivot(index='date', columns='type', values='fare')
sum_fares_pivot
```

### DataFrame using Loc() 

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
set_dates_df = sum_fares_pivot.loc['2019-01-01':'2019-04-29']
set_dates_df
```

### DataFrame using resample()

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
set_dates_df.index = pd.to_datetime(set_dates_df.index)

```

```python
# 7. Check that the datatype for the index is datetime using df.info()
set_dates_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
fare_sum_by_week_df = set_dates_df.resample("W").sum()
fare_sum_by_week_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style

# Use the graph style fivethirtyeight.


# plot

fare_sum_by_week_df.plot(figsize=(12,4))
plt.xlabel('')
plt.ylabel('Fare ($USD)')
plt.title('Total Fare by City Type')
plt.style.use('fivethirtyeight')
plt.legend(loc='best')
```
